```python
import pandas as pd
import plotly_express as px
```

```python
df = pd.read_csv("../my_gurgaon_properties_missing_value_imputation.csv")
latlong = pd.read_csv("datasets/latlong.csv")
```

```python
latlong['latitude'] = latlong['coordinates'].str.split(',').str.get(0).str.split('°').str.get(0).astype('float')
latlong['longitude'] = latlong['coordinates'].str.split(',').str.get(1).str.split('°').str.get(0).astype('float')
```

```python
new_df = df.merge(latlong, on='sector')
```

```python
group_df = new_df.groupby("sector").mean(numeric_only=True)[['price','price_per_sqft','built_up_area','latitude','longitude']]
group_df
```

```python
# fig = px.scatter_mapbox(group_df, lat="latitude", lon="longitude", color="price_per_sqft", size="built_up_area", color_continuous_scale=px.colors.cyclical.IceFire, zoom=10, mapbox_style="open-street-map", text=group_df.index)

fig = px.scatter_map(group_df, lat="latitude", lon="longitude", color="price_per_sqft", size="built_up_area", color_continuous_scale=px.colors.cyclical.IceFire, zoom=10, map_style="open-street-map", text=group_df.index)
fig.show()
```

```python
new_df.to_csv("dataviz_1.csv", index=False)
```
